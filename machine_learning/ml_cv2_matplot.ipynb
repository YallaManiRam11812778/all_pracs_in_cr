```python
# displaying RGB channels for our images
figure , axis = plt.subplots(1,3, figsize=(10,10))
print(axis,"-----------",len(axis),"================",type(axis))
axis[0].imshow(plt_reading_image[:,:,0],cmap="Reds")
axis[1].imshow(plt_reading_image[:,:,1],cmap="Greens")
axis[2].imshow(plt_reading_image[:,:,2],cmap="Blues")
axis[0].axis("off")
axis[0].set_title("RED")

axis[1].axis("off")
axis[1].set_title("Green")

axis[2].axis("off")
axis[2].set_title("BLUE")
```

```python
# Dieeference between CV2 And Matplotlib
fig , axis = plt.subplots(1,2,figsize=(10,10))
axis[1].imshow(plt_reading_image)
axis[1].set_title("PLT")
axis[0].imshow(cv2_reading_image)
axis[0].set_title("CV2")
```
