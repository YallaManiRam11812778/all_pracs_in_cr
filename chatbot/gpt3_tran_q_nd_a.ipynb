```python
from transformers import pipeline

# Load the Q&A pipeline
qa_pipeline = pipeline("question-answering")
```

```python
# context = """
# ChatGPT is an AI language model developed by OpenAI, based on the GPT-4 architecture.
# It is designed to generate human-like text based on the input it receives. 
# It has a wide range of applications, including answering questions, writing essays, and creating conversational agents.
# """
context = """
my name is Ram from Caratred Technologies LLP with annual income of 3.75 LPA as python developer.
I live in hostel in Suchitra. I like Biriyani the most and that too only Hyderabadi Biriyani is my most favorite.
My bro works along with me and his name is Lksmn.
languages that I know are:  python, js, html, java and the count is 5
"""
```

```python
questions = [
    "what is my name?",
    "occupation?",
    "which dish do you prefer?",
    "what is your salary?",
    "bhai ka naam?",
    "how many languages that u know?"
]

for question in questions:
    result = qa_pipeline(question=question, context=context)
    print(result)
    print(f"Question: {question}")
    print(f"Answer: {result['answer']}\n")
```
